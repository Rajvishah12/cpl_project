```python
#importing library data. MAKE SURE YOU ARE IMPORTING AS A GEOPANDAS!
branchLocations = gpd.read_file('../data/clean/branches.csv')
```

```python
branchLocations.crs = "EPSG:4326"
```

```python
branchLocations.crs
```

```python
#Replacing Location with longitude and latitude columns
branchLocations[['LATITUDE', 'LONGITUDE']] = [ast.literal_eval(x)[:2] for x in branchLocations['LOCATION']]
#Converting latitude and longitude to numeric types
branchLocations.loc[:, 'LATITUDE'] = pd.to_numeric(branchLocations.loc[:, 'LATITUDE'])
branchLocations.loc[:, 'LONGITUDE'] = pd.to_numeric(branchLocations.loc[:, 'LONGITUDE'])
#checking new columns
branchLocations.head()
```

```python
#assigning isochrone polygons to the geometry column
for i in range(0, len(branchLocations['geometry'])):
    api_response =  api.get_data((branchLocations.loc[i,'LATITUDE'],branchLocations.loc[i,'LONGITUDE']), 15)

    branchLocations.loc[i,'geometry'] = shape(api_response)
```

```python
#reassigning to a column that better describes the data
branchLocations['15 min drive'] = branchLocations['geometry']
```

```python
#checking new column
branchLocations.head()
```

```python
#This function will give the isochrone geometry for a point given a data frame, the geometry, longitude, and latitude column name as strings,
#the name of your api, time you'd like to calculate, an the name of the new column of information as a string.
def getIsochrone(dataFrame, geoColName, lonColName, latColName, apiName, time, descripColName):
    for i in range(0, len(dataFrame[geoColName])):
     api_response =  apiName.get_data((dataFrame.loc[i, latColName],branchLocations.loc[i,lonColName]), time)
     dataFrame.loc[i, geoColName] = shape(api_response)
    
    dataFrame[descripColName] = dataFrame[geoColName]
    return dataFrame
```

```python
#example of function
testFrame= getIsochrone(branchLocations, 'geometry', 'LONGITUDE', 'LATITUDE', api, 5, '5 min drive')
```
