```python
import pandas as pd
```

```python
libraryBranchInfo2019_2021 = pd.read_csv('../../data/final_data/libInfo2019_2021.csv')
#walking20Demographics = pd.read_csv('../../data/final_data/walking20Demographics.csv')
avgWalkTimeNoLib = pd.read_csv('../../data/final_data/averageWalkingTimeNeighborhoodsNoLibIncluded.csv')
```

```python
libInfo2019_2021 = libraryBranchInfo2019_2021.drop(columns=['Unnamed: 0']).rename(columns= {"BRANCH":"Branch"})
libInfo2019_2021
```

```python

nbhdBranchDemosNoLib = avgWalkTimeNoLib.drop(columns=['Unnamed: 0'])
nbhdBranchDemosNoLib['Branch'] = nbhdBranchDemosNoLib['Branch'].replace(['Woodson Regional Library'], 'Woodson Regional')
nbhdBranchDemosNoLib
```
