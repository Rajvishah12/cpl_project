## Reading in CSV's

```python
import pandas as pd
```

```python
atOrAbovePoverty = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/AtorAbovePovertyAverageWalkingNeighborhoods.csv')
inPoverty = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/LivinginPovertyAverageWalkingNeighborhoods.csv')
highSchoolOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HighSchoolorMoreAverageWalkingNeighborhoods.csv')
someCollegeOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/SomeCollegeorMoreAverageWalkingNeighborhoods.csv')
employed = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/EmployedAverageWalkingNeighborhoods.csv')
unemployed = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/UnemployedAverageWalkingNeighborhoods.csv')
transpCarTruckVan = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/TransportationCarTruckVanAverageWalkingNeighborhoods.csv')
white = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/WhiteAloneaverageWalkingTimeNeighborhoods.csv')
blackOrAfricanAmerican = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/BlackorAfricanAmericanAloneaverageWalkingTimeNeighborhoods.csv')
hasInternet = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HasInternetAverageWalkingNeighborhoods.csv')
noInternet = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/NoInternetAverageWalkingNeighborhoods.csv')
inc150000to199999 = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/Income$150,000to199,999AverageWalkingNeighborhoods.csv')
bachelorsOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/BachelorsDegreeorMoreAverageWalkingNeighborhoods.csv')
mastersOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/MastersDegreeorMoreAverageWalkingNeighborhoods.csv')
hasComputer = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HasComputersAverageWalkingNeighborhoods.csv')
noComputer = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/NoComputersAverageWalkingNeighborhoods.csv')
```

```python
white = white.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'White'})
blackOrAfricanAmerican = blackOrAfricanAmerican.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Black or African American'})

highSchoolOrMore = highSchoolOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Degree: High School or More'})
someCollegeOrMore = someCollegeOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Degree: Some College or More'})
bachelorsOrMore = bachelorsOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Degree: Bachelors or More'})
mastersOrMore = mastersOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Degree: Masters or More'})

atOrAbovePoverty = atOrAbovePoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'At or Above Poverty'})
inPoverty = inPoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'In Poverty'})
inc150000to199999 = inc150000to199999.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Income: $150,000 to $199,999'})
employed = employed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Employed'})
unemployed = unemployed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Unemployed'})

hasInternet = hasInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has Internet Access'})
noInternet = noInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has No Internet Access'})
hasComputer = hasComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has Computer'})
noComputer = noComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has No Computer'})

transpCarTruckVan = transpCarTruckVan.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Transportation: Car, Truck, or Van '})
```

```python
first_merge = white.merge(blackOrAfricanAmerican,on='Branch') 

second_merge = first_merge.merge(highSchoolOrMore,on='Branch') 

third_merge = second_merge.merge(someCollegeOrMore,on='Branch') 

fourth_merge = third_merge.merge(bachelorsOrMore,on='Branch') 

fifth_merge = fourth_merge.merge(mastersOrMore,on='Branch') 

sixth_merge = fifth_merge.merge(atOrAbovePoverty,on='Branch') 

seventh_merge = sixth_merge.merge(inPoverty,on='Branch') 

eighth_merge = seventh_merge.merge(inc150000to199999,on='Branch') 

ninth_merge = eighth_merge.merge(employed,on='Branch') 

tenth_merge = ninth_merge.merge(unemployed,on='Branch') 

eleventh_merge = tenth_merge.merge(hasInternet,on='Branch') 

twelfth_merge = eleventh_merge.merge(noInternet,on='Branch') 

thirteenth_merge = twelfth_merge.merge(hasComputer,on='Branch') 

fourteenth_merge = thirteenth_merge.merge(noComputer,on='Branch') 

fifteenth_merge = fourteenth_merge.merge(transpCarTruckVan,on='Branch') 

fifteenth_merge
```

```python
fifteenth_merge.to_csv('../../notebooks/CSV Compile for Library Info and Nbhs/demosDFAverageWalkingNeighborhood.csv')
```
