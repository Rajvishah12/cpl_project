## Reading in CSV's

```python
import pandas as pd
```

```python
atOrAbovePoverty = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/AtorAbovePovertyAverageWalkingNeighborhoods.csv')
inPoverty = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/LivinginPovertyAverageWalkingNeighborhoods.csv')
highSchoolOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HighSchoolorMoreAverageWalkingNeighborhoods.csv')
someCollegeOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/SomeCollegeorMoreAverageWalkingNeighborhoods.csv')
employed = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/EmployedAverageWalkingNeighborhoods.csv')
unemployed = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/UnemployedAverageWalkingNeighborhoods.csv')
transpCarTruckVan = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/TransportationCarTruckVanAverageWalkingNeighborhoods.csv')
white = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/WhiteAloneaverageWalkingTimeNeighborhoods.csv')
blackOrAfricanAmerican = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/BlackorAfricanAmericanAloneaverageWalkingTimeNeighborhoods.csv')
hasInternet = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HasInternetAverageWalkingNeighborhoods.csv')
noInternet = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/NoInternetAverageWalkingNeighborhoods.csv')

incLessthan10000 = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/IncomeLessThan$10,000.csv')
inc10000to14999 = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/Income$10,000to$14,999.csv')
inc25000to29999 = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/Income$25,000to$29,999.csv')
inc150000to199999 = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/Income$150,000to199,999AverageWalkingNeighborhoods.csv')


bachelorsOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/BachelorsDegreeorMoreAverageWalkingNeighborhoods.csv')
mastersOrMore = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/MastersDegreeorMoreAverageWalkingNeighborhoods.csv')
hasComputer = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/HasComputersAverageWalkingNeighborhoods.csv')
noComputer = pd.read_csv('../CSV Compile for Library Info and Nbhs/demoCSVAverageWalkingNeighborhood/NoComputersAverageWalkingNeighborhoods.csv')
```

```python
white = white.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: White'})
blackOrAfricanAmerican = blackOrAfricanAmerican.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Black or African American'})

highSchoolOrMore = highSchoolOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Degree: High School or More'})
someCollegeOrMore = someCollegeOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Degree: Some College or More'})
bachelorsOrMore = bachelorsOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Degree: Bachelors or More'})
mastersOrMore = mastersOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Degree: Masters or More'})

atOrAbovePoverty = atOrAbovePoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: At or Above Poverty'})
inPoverty = inPoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: In Poverty'})

incLessthan10000 = incLessthan10000.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Income: Less than 10,000'})
inc10000to14999 = inc10000to14999.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Income: $10,000 to $14,999'})
inc25000to29999 = inc25000to29999.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Income: $25,000 to $29,999'})
inc150000to199999 = inc150000to199999.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Income: $150,000 to $199,999'})


employed = employed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Employed'})
unemployed = unemployed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Unemployed'})

hasInternet = hasInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Has Internet Access'})
noInternet = noInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Has No Internet Access'})
hasComputer = hasComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Has Computer'})
noComputer = noComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Has No Computer'})

transpCarTruckVan = transpCarTruckVan.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'WT: Transportation: Car, Truck, or Van '})
```

```python
first_merge = white.merge(blackOrAfricanAmerican,on='Branch') 

second_merge = first_merge.merge(highSchoolOrMore,on='Branch') 

third_merge = second_merge.merge(someCollegeOrMore,on='Branch') 

fourth_merge = third_merge.merge(bachelorsOrMore,on='Branch') 

fifth_merge = fourth_merge.merge(mastersOrMore,on='Branch') 

sixth_merge = fifth_merge.merge(atOrAbovePoverty,on='Branch') 

seventh_merge = sixth_merge.merge(inPoverty,on='Branch') 

eighth_merge = seventh_merge.merge(incLessthan10000,on='Branch') 

ninth_merge = eighth_merge.merge(inc10000to14999,on='Branch') 

tenth_merge = ninth_merge.merge(inc25000to29999,on='Branch') 

eleventh_merge = tenth_merge.merge(inc150000to199999,on='Branch') 

twelfth_merge = eleventh_merge.merge(employed,on='Branch') 

thirteenth_merge = twelfth_merge.merge(unemployed,on='Branch') 

fourteenth_merge = thirteenth_merge.merge(hasInternet,on='Branch') 

fifteenth_merge = fourteenth_merge.merge(noInternet,on='Branch') 

sixteenth_merge = fifteenth_merge.merge(hasComputer, on='Branch')

seventeenth_merge = sixteenth_merge.merge(noComputer, on='Branch')

eighteenth_merge = seventeenth_merge.merge(transpCarTruckVan, on='Branch')

eighteenth_merge
```

```python
eighteenth_merge.to_csv('../../notebooks/CSV Compile for Library Info and Nbhs/demosDFAverageWalkingNeighborhood.csv')
```
