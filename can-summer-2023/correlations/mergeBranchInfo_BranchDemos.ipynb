```python
import pandas as pd
```

```python
libraryBranchInfo2019_2021 = pd.read_csv('../data/dani data/libInfo2019_2021.csv')
cycling20Demographics = pd.read_csv('../library neighborhoods/avgCyclingTimeNeighborhoodsDems.csv')
```

```python
libInfo2019_2021 = libraryBranchInfo2019_2021.drop(columns=['Unnamed: 0']).rename(columns= {"BRANCH":"Branch"})
libInfo2019_2021.head(3)
```

```python
cycling20Demographics
```

```python
libInfo2019_2021['AVG WALKING TIME (MIN)'] = cycling20Demographics['Average Cycling Time (Min)']
```

```python
libInfo2019_2021.head(3)
```

```python
libInfo2019_2021 = libInfo2019_2021.rename(columns={"AVG WALKING TIME (MIN)": "AVG CYCLING TIME (MIN)"})
```

```python
neighborhoodBranchDemos = cycling20Demographics.drop(columns=['Unnamed: 0'])
neighborhoodBranchDemos['Branch'] = neighborhoodBranchDemos['Branch'].replace(['Woodson Regional Library'], 'Woodson Regional')
neighborhoodBranchDemos
```

```python
neighborhoodBranchInfo_Demos = libInfo2019_2021.merge(neighborhoodBranchDemos, on='Branch')
neighborhoodBranchInfo_Demos
```

```python
neighborhoodBranchInfo_Demos = neighborhoodBranchInfo_Demos.drop(columns = ['Average Cycling Time (Min)'])
```

```python
neighborhoodBranchInfo_Demos.to_csv('../data/dani data/cyclingNeighborhoodBranchInfo_Demos.csv')
```
