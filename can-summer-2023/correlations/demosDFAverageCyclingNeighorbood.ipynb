## Reading in CSV's

```python
import pandas as pd
```

```python
atOrAbovePoverty = pd.read_csv('../library neighborhoods/avgTravelTime/abovePovertyCyclingTimeNeighborhoods.csv')
inPoverty = pd.read_csv('../library neighborhoods/avgTravelTime/inPovertyCyclingTimeNeighborhoods.csv')
highSchoolOrMore = pd.read_csv('../library neighborhoods/avgTravelTime/atLeastHighSchoolCyclingTimeNeighborhoods.csv')
bachelorsOrMore = pd.read_csv('../library neighborhoods/avgTravelTime/bachelorsCyclingTimeNeighborhoods.csv')
employed = pd.read_csv('../library neighborhoods/avgTravelTime/employedCyclingTimeNeighborhoods.csv')
unemployed = pd.read_csv('../library neighborhoods/avgTravelTime/unemployedCyclingTimeNeighborhoods.csv')
#transpCarTruckVan = pd.read_csv('../CSV Compile for Library Info and Nbhs/
#demoCSVAverageWalkingNeighborhood/TransportationCarTruckVanAverageWalkingNeighborhoods.csv')
white = pd.read_csv('../library neighborhoods/avgTravelTime/whiteCyclingTimeNeighborhoods.csv')
blackOrAfricanAmerican = pd.read_csv('../library neighborhoods/avgTravelTime/blackCyclingTimeNeighborhoods.csv')
hasInternet = pd.read_csv('../library neighborhoods/avgTravelTime/withInternetCyclingTimeNeighborhoods.csv')
noInternet = pd.read_csv('../library neighborhoods/avgTravelTime/withoutInternetCyclingTimeNeighborhoods.csv')
inc150000to199999 = pd.read_csv('../library neighborhoods/avgTravelTime/income150kTo199kCyclingTimeNeighborhoods.csv')
mastersOrMore = pd.read_csv('../library neighborhoods/avgTravelTime/mastersCyclingTimeNeighborhoods.csv')
# NO FILE hasComputer = pd.read_csv('../library neighborhoods/avgTravelTime/mastersCyclingTimeNeighborhoods.csv')
# NO FILE noComputer = pd.read_csv('../library neighborhoods/avgTravelTime/mastersCyclingTimeNeighborhoods.csv')
```

```python
white = white.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling White'})
blackOrAfricanAmerican = blackOrAfricanAmerican.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Black or African American'})

highSchoolOrMore = highSchoolOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling: High School or More'})
#someCollegeOrMore = someCollegeOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Degree: Some College or More'})
bachelorsOrMore = bachelorsOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling: Bachelors or More'})
mastersOrMore = mastersOrMore.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling: Masters or More'})

atOrAbovePoverty = atOrAbovePoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling At or Above Poverty'})
inPoverty = inPoverty.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling In Poverty'})
inc150000to199999 = inc150000to199999.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Income: $150,000 to $199,999'})
employed = employed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Employed'})
unemployed = unemployed.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Unemployed'})

hasInternet = hasInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Has Internet Access'})
noInternet = noInternet.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Cycling Time (Minutes)': 'Avg Cycling Has No Internet Access'})
#hasComputer = hasComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has Computer'})
#noComputer = noComputer.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Has No Computer'})

#transpCarTruckVan = transpCarTruckVan.drop(columns = ['Unnamed: 0']).set_index('Branch').rename(columns={'Average Walking Time (Minutes)': 'Transportation: Car, Truck, or Van '})
```

```python
first_merge = white.merge(blackOrAfricanAmerican,on='Branch') 

second_merge = first_merge.merge(highSchoolOrMore,on='Branch') 

#third_merge = second_merge.merge(someCollegeOrMore,on='Branch') 

fourth_merge = second_merge.merge(bachelorsOrMore,on='Branch') 

fifth_merge = fourth_merge.merge(mastersOrMore,on='Branch') 

sixth_merge = fifth_merge.merge(atOrAbovePoverty,on='Branch') 

seventh_merge = sixth_merge.merge(inPoverty,on='Branch') 

eighth_merge = seventh_merge.merge(inc150000to199999,on='Branch') 

ninth_merge = eighth_merge.merge(employed,on='Branch') 

tenth_merge = ninth_merge.merge(unemployed,on='Branch') 

eleventh_merge = tenth_merge.merge(hasInternet,on='Branch') 

twelfth_merge = eleventh_merge.merge(noInternet,on='Branch') 

#thirteenth_merge = twelfth_merge.merge(hasComputer,on='Branch') 

#fourteenth_merge = thirteenth_merge.merge(noComputer,on='Branch') 

#fifteenth_merge = fourteenth_merge.merge(transpCarTruckVan,on='Branch') 

twelfth_merge
```

```python
twelfth_merge.to_csv('../library neighborhoods/avgTravelTime/allAvgCyclingTimes.csv')
```
